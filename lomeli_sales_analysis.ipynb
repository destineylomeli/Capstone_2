## Question 3: Store Sales Performance Rankings
This section ranks store performance within the Maryland and Massachusetts territories based on total sales revenue.

```python
store_performance = territory_sales.groupby(
    ['State', 'Store ID']
)['Sale Amount'].sum().reset_index
```

```python
store_performance = store_performance.sort_values(
    by='Sale Amount',
    ascending=False
)
```

```python
store_performance_display = store_performance.copy()

store_performance_display['Sale Amount'] = (
    store_performance_display['Sale Amount']
    .round(2)
    .apply(lambda x: f'${x:,.2f}')
)
store_performance_display
```
