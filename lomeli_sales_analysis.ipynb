```python
northeast_territory = store_detail[
    store_detail['State'].isin(['Maryland', 'Massachusetts'])
    ]

northeast_territory
```

## Question 2: Monthly Total Revenue
This section analyzes monthly in store sales revenue for stores located in Maryland and Massachusetts within the Northeast region.

```python
monthly_revenue = store_sales.merge(
    northeast_territory,
    on='Store ID',
    how='left'
)

monthly_revenue['Transaction Date'] = pd.to_datetime(
    monthly_revenue['Transaction Date']
)

monthly_revenue['Month'] = (
    monthly_revenue['Transaction Date']
    .dt.to_period('M')
)

monthly_revenue = monthly_revenue.groupby(
    ['State', 'Month']
)['Sale Amount'].sum().reset_index()
```

```python
monthly_revenue_display = monthly_revenue.copy()

monthly_revenue_display['Sale Amount'] = monthly_revenue_display['Sale Amount'].map(
    '${:,.2f}'.format
)
```

```python
monthly_revenue_display[
    monthly_revenue_display['State'] == 'Maryland'
    ]
```

```python
monthly_revenue_display[
    monthly_revenue_display['State'] == 'Massachusetts'
    ]
```

```python
monthly_chart = monthly_revenue.copy()

monthly_chart['Month'] = monthly_chart['Month'].astype(str)

monthly_chart['Sale Amount'] = (
    monthly_chart['Sale Amount']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

import matplotlib.pyplot as plt

plt.figure(figsize=(14,5))

for state in monthly_chart['State'].unique():
    
    state_data = monthly_chart[
        monthly_chart['State'] == state
        ]
    
    plt.plot(
        state_data['Month'].values,
        state_data['Sale Amount'].values,
        marker='o',
        label=state
    )

plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue Trends by State')

plt.legend()
plt.xticks(
    monthly_chart['Month'][::3],
    rotation=45
)

plt.tight_layout()
plt.show()
```

## Question 4: Top Customers
This section identifies the top customers in the Maryland and Massachusetts territories by comparing RewardsID from the sales data with customer IDs from the customer list data.

## Question 5: Product Category Performance
This section analyzes product category sales performance in Maryland and Massachusetts to identify the highest revenue generating categories.

```python
# Merge sales with store details to get State
q5_data = store_sales.merge(
    northeast_territory,
    on='Store ID',
    how='left'
)

# Merge sales with products to get CategoryID
q5_data = q5_data.merge(
    products[['Prod Num', 'CategoryID']],
    on='Prod Num',
    how='left'
)

# Merge with product categories to get category names
q5_data = q5_data.merge(
    product_categories[['CategoryID', 'Category']],
    on='CategoryID',
    how='left'
)

# Convert transaction date to month
q5_data['Transaction Date'] = pd.to_datetime(q5_data['Transaction Date'])
q5_data['Month'] = q5_data['Transaction Date'].dt.to_period('M')

# Count transactions per month by state and category
transaction_count = (
    q5_data
    .groupby(['State', 'Month', 'Category'])
    .size()
    .reset_index(name='Transaction Count')
)
```
