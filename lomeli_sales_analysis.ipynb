## Question 1: Territory Managers and Store Locations
The assigned sales territory is located in the Northeast region. 

The assigned territories include:

- Massachusetts

This analysis will focus on stores located within these territories to evaluate sales performance, customer activity, product category trends, and future marketing opportunities.

```python
northeast_territory = store_detail[
    store_detail['State'].isin(['Maryland', 'Massachusetts'])
    ]

northeast_territory
```

## Question 2: Monthly Total Revenue
This section analyzes monthly in store sales revenue for stores located in Maryland and Massachusetts within the Northeast region.

```python
monthly_revenue = store_sales.merge(
    northeast_territory,
    on='Store ID',
    how='left'
)

monthly_revenue['Transaction Date'] = pd.to_datetime(
    monthly_revenue['Transaction Date']
)

monthly_revenue['Month'] = (
    monthly_revenue['Transaction Date']
    .dt.to_period('M')
)

monthly_revenue = monthly_revenue.groupby(
    ['State', 'Month']
)['Sale Amount'].sum().reset_index()
```

```python
monthly_revenue_display = monthly_revenue.copy()

monthly_revenue_display['Sale Amount'] = monthly_revenue_display['Sale Amount'].map(
    '${:,.2f}'.format
)
```

```python
monthly_revenue_display[
    monthly_revenue_display['State'] == 'Maryland'
    ]
```

```python
monthly_revenue_display[
    monthly_revenue_display['State'] == 'Massachusetts'
    ]
```

```python
monthly_chart = monthly_revenue.copy()

monthly_chart['Month'] = monthly_chart['Month'].astype(str)

monthly_chart['Sale Amount'] = (
    monthly_chart['Sale Amount']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

import matplotlib.pyplot as plt

plt.figure(figsize=(14,5))

for state in monthly_chart['State'].unique():
    
    state_data = monthly_chart[
        monthly_chart['State'] == state
        ]
    
    plt.plot(
        state_data['Month'].values,
        state_data['Sale Amount'].values,
        marker='o',
        label=state
    )

plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue Trends by State')

plt.legend()
plt.xticks(
    monthly_chart['Month'][::3],
    rotation=45
)

plt.tight_layout()
plt.show()
```

## Question 3: Store Sales Performance Rankings
This section ranks store performance within the Maryland and Massachusetts territories based on total sales revenue.

```python
store_performance = territory_sales.groupby(
    ['State', 'Store ID']
)['Sale Amount'].sum().reset_index()
```

```python
store_performance = store_performance.merge(
    northeast_territory[['Store ID', 'Store Location']],
    on='Store ID',
    how='left'
)
```

```python
store_performance = store_performance[
    ['Store ID', 'State', 'Store Location', 'Sale Amount']
]

store_performance = store_performance.sort_values(
    by='Sale Amount',
    ascending=False
)
```

```python
store_performance_display = store_performance.copy()

store_performance_display['Sale Amount'] = (
    store_performance_display['Sale Amount']
    .round(2)
    .apply(lambda x: f'${x:,.2f}')
)
```

```python
store_performance_display[
    store_performance_display['State'] == 'Maryland'
    ]
```

```python
store_performance_display[
    store_performance_display['State'] == 'Massachusetts'
    ]
```

## Question 4: Top Customers
This section identifies the top customers in the Maryland and Massachusetts territories by comparing RewardsID from the sales data with customer IDs from the customer list data.

```python
top_customers = territory_sales.groupby(
    ['State', 'RewardsID']
)['Sale Amount'].sum().reset_index()

top_customers = top_customers.dropna(subset=['RewardsID'])

top_customers['RewardsID'] = top_customers['RewardsID'].astype(int)
```

```python
top_customers = top_customers.merge(
    customer_list[['cust_id', 'name', 'email']],
    left_on='RewardsID',
    right_on='cust_id',
    how='left'
)
```

```python
top_customers = top_customers.sort_values(
    by='Sale Amount',
    ascending=False
)
```

```python
top_customers_display = top_customers.copy()

top_customers_display['Sale Amount'] = (
    top_customers_display['Sale Amount']
    .round(2)
    .apply(lambda x: f'${x:,.2f}')
    )
```

```python
top_customers_display[
    top_customers_display['State'] == 'Maryland'
    ][['State', 'RewardsID', 'name', 'email', 'Sale Amount']].head(10)
```

```python
top_customers_display[
    top_customers_display['State'] == 'Massachusetts'
    ][['State', 'RewardsID', 'name', 'email', 'Sale Amount']].head(10)
```

## Question 5: Product Category Performance
This section analyzes product category sales performance in Maryland and Massachusetts to identify the highest revenue generating categories.

```python
category_data = territory_sales.merge(
    northeast_territory[['Store ID', 'Store Location']],
    on='Store ID', 
    how='left'
)
```

```python
category_sales_display = (
    category_data
    .groupby(['State', 'Store Location_x', 'CategoryID_x'])['Sale Amount']
    .sum()
    .reset_index()
    .merge(
        product_categories[['CategoryID', 'Category']],
        left_on='CategoryID_x',
        right_on='CategoryID',
        how='left'
    )
    .rename(columns={
        'Store Location_x': 'Store Location', 
        'CategoryID_x': 'CategoryID'
    })
)
category_sales_display['Sale Amount'] = category_sales_display['Sale Amount'].apply(
    lambda x: f'${x:,.2f}'
)
```

```python
maryland_top_categories = (
    category_sales_display[
        category_sales_display['State'] == 'Maryland'
        ]
    [['State', 'Category', 'Sale Amount']]
    .groupby(['State', 'Category'])['Sale Amount']
    .first()
    .reset_index()
    .sort_values(by='Sale Amount', ascending=False)
)

maryland_top_categories
```

```python
maryland_chart = maryland_top_categories.copy()

maryland_chart['Revenue'] = (
    maryland_chart['Sale Amount']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

maryland_chart = maryland_chart.sort_values(
    by='Revenue',
    ascending=True
)

plt.figure(figsize=(8,5))
plt.barh(maryland_chart['Category'], maryland_chart['Revenue'])

plt.xlabel('Revenue')
plt.ylabel('Category')
plt.title('Top Product Categories in Maryland')

plt.tight_layout()
plt.show()
```

```python
massachusetts_top_categories = (
    category_sales_display[
        category_sales_display['State'] == 'Massachusetts'
        ]
    [['State', 'Category', 'Sale Amount']]
    .groupby(['State', 'Category'])['Sale Amount']
    .first()
    .reset_index()
    .sort_values(by='Sale Amount', ascending=False)
)

massachusetts_top_categories
```

```python
massachusetts_chart = massachusetts_top_categories.copy()

massachusetts_chart['Revenue'] = (
    massachusetts_chart['Sale Amount']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

massachusetts_chart = massachusetts_chart.sort_values(
    by='Revenue',
    ascending=True
)

plt.figure(figsize=(8,5))
plt.barh(massachusetts_chart['Category'], massachusetts_chart['Revenue'])

plt.xlabel('Revenue')
plt.ylabel('Category')
plt.title('Top Product Categories in Massachusetts')

plt.tight_layout()
plt.show()
```
